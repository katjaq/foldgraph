```python
import os
import sys
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

The skeleton of a baboon neocortical folding pattern was extracted, and from there we created a simplified version: the foldgraph. This graph is in the file `baboon_polar_graph.txt`.

```python
# Read a fold graph
fg=!cat baboon_polar_graph.txt
nv,_,ne = [int(x) for x in fg[0].split(" ")]
v = np.array([[float(y) for y in x.split(" ")] for x in fg[1:nv+1]])[:,:2]
e = np.array([[int(y) for y in x.split(" ")] for x in fg[1+nv:-1]])
```

```python
# Make a networkx graph and plot
G = nx.Graph()
G.add_edges_from(e)
print("Number of nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())

plt.figure()
nx.draw(G, v, with_labels=True)
```

The left and right hemispheres of the baboon brain have a very similar folding pattern. Their foldgraphs are independent. Here we extract each hemisphere as an independent graph.

```python
C = nx.connected_components(G)
G1 = G.subgraph(next(C))
G2 = G.subgraph(next(C))
```

```python
plt.figure()
plt.subplot(121)
nx.draw(G1, v, with_labels=True)
plt.subplot(122)
nx.draw(G2, v, with_labels=True)
```
